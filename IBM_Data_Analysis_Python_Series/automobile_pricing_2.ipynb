<H2>Data Normalisation</H2

__Defintion__ â€“Â It is the process of transforming the values of several variables into a similar range. 

Typical normalisations include: 
* Scaling the variable so the variable average is 0
* Scaling the variable so the variance is 1
* Scaling the variable so the variable values range from 0 to 1

<b>Example</b>
<p>To demonstrate normalization, I will scale the columns "length", "width" and "height".</p>
<p><b>Target:</b> normalize those variables so their value ranges from 0 to 1</p>
<p><b>Approach:</b> replace the original value by (original value)/(maximum value)</p>

```python
## Before transformation: 
print(df1[['length', 'width']])
```

```python
# After transformation: Replace (original value) by the (original value/max value)
df1['length'] = df1['length']/df1['length'].max()
df1['width'] = df1['width']/df1['width'].max()
print(df1[['length', 'width']])
```

Normalization of the column `'height'`

```python
# Before transformation: 
print(df1[['height']])
```

```python
# After transformation: Replace the column of interest (original value) with the formula of (original value/max value)
df1['height'] = df1['height']/df1['height'].max()
print(df1[['height']])
```

```python
# Final outcome of all the transformations: 
df1[['length', 'width', 'height']].head()
```

<h2>Binning in Pandas</h2>

__Definition__ â€“Â It is a process of tranforming continuous numerical variable into discrete categorical `'bins'` for grouped analysis.

Using Pandas, I will cut the segment of `'horsepower'` into 3 segments (`bins`).

```python
df1[['horsepower']].head()
```

Rearrangement of `'horsepower'` into 3 bins to compare them against the pricing of the car.

```python
# Histogram plot of the distribution of the horsepower
%matplotlib inline 
from matplotlib import pyplot
plt.pyplot.hist(df['horsepower'])
plt.pyplot.hist(df['horsepower'])

# Set x/y labels and plot title
plt.pyplot.xlabel('Horsepower')
plt.pyplot.ylabel('Count')
plt.pyplot.title('Horsepower bins')
```

Bin distribution into 3 compoartments â€“Â 

`linspace(start_value, end_value, number_generated)`

I will segregate the `horsepower` into 3 equal sized bins: 

<p>start_value = min(df["horsepower"]).</p>
<p>end_value = max(df["horsepower"]).</p>
<p>4 dividers to divide into 3 equal compartments, so numbers_generated = 4</p>

```python
bins = np.linspace(min(df1['horsepower']), max(df1['horsepower']), 4)
bins
```

```python
# Set group names of the bins
grp_name = ['Low', "Medium", 'High']
```

```python
# I will now apply the 'cut' function to determine what each value of df['horsepower'] belongs to: 
df1['horsepower-binned'] = pd.cut(df1['horsepower'], bins, labels = grp_name, include_lowest = True)
df1[['horsepower', 'horsepower-binned']].head(20)
```

```python
# Number of vehicles in each bin: 
df1['horsepower-binned'].value_counts()
```

```python
# Bar-Plot of the above distribution into each bin: 
pyplot.bar(grp_name, df1['horsepower-binned'].value_counts())

# Set x/y labels and plot title: 
plt.pyplot.xlabel('Horsepower')
plt.pyplot.ylabel('Count')
plt.pyplot.title("Horsepower Bins")
```

<h2>Indicator Variables</h2>

__Definition__ â€“Â It is a numerical variable ("dummy" variable) used to label categories. They are called "dummies" because the numbers themselves don't have inherent meaning. 

It converts the categorical data and enables it to undergo `regression analysis`. 
* Regression analysis only understands numeric values.
* It does not understand words/strings/objects

```python
df.columns
```

```python
# I will use the `get_dummies()` from the Pandas library to convert the object type categorical variable which will enable it to undergo regression analysis. 
dummy_variable_1 = pd.get_dummies(df1['fuel-type'])
dummy_variable_1.head()
```

```python
# Column names changed for clarity: 
dummy_variable_1.rename(columns = {'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace = True)
dummy_variable_1.head()
```

```python
# Merging the data frame df1 with the newly formed "dummy_variable_1"
df1 = pd.concat([df1, dummy_variable_1], axis=1)

# Drop original column "fuel-type" from "df"
df1.drop("fuel-type", axis = 1, inplace=True)
```
