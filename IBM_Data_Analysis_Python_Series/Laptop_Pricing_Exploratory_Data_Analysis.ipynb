<h3>Task_3 – GroupBy and Pivot Tables</h3>

Group the parameters "GPU", "CPU_core" and "Price" to make a pivot table and visualize this connection using the pcolor plot.


```python
# Creation of the group: 
df_gptest_1 = df[['GPU', 'CPU_core', 'Price']]
grouped_test1 = df_gptest_1.groupby(['GPU', 'CPU_core'], as_index = False).mean()
print(grouped_test1)
```

```python
# Creation of the pivot table of the above groups: 
grouped_pivot_1 = grouped_test1.pivot(index='GPU', columns='CPU_core')
grouped_pivot_1
```

```python
# Create the plot: 
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot_1, cmap='RdBu')

#label names
row_labels = grouped_pivot_1.columns.levels[1]
col_labels = grouped_pivot_1.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot_1.shape[1]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

fig.colorbar(im)
```

### Task 4 - Pearson Correlation and p-values

Use the `scipy.stats.pearsonr()` function to evaluate the Pearson Coefficient and the p-values for each parameter tested above. This will help in determining the parameters most likely to have a strong effect on the price of the laptops.

```python
# Use the For-loop to iterate over each attribute with respect to price and display the result:
for param in ['RAM_GB', 'CPU_frequency', 'Storage_GB_SSD', 'Screen_Size_inch', 'Weight_pounds', 'CPU_core', 'OS', 'GPU', 'Category']:
    pearson_coef, p_value = stats.pearsonr(df[param], df['Price'])
    print(param)
    print("–– The Pearson Correlation coefficient for", param, 'is', pearson_coef, 'with a p-value of p =', p_value)
```
